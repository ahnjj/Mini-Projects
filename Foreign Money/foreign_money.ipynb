```python
# url 에서 데이터를 가져와 문자열로 반환
import json, requests
import pandas as pd
url = requests.get("https://www.koreaexim.go.kr/site/program/financial/exchangeJSON?authkey=7C2WbdJeR4Jugo687WwfvKNckPyU5HdD&searchdate=20230802&data=AP01")
data = url.json()

# 나라이름 : 환율
# ctr_cur = { data[i]['cur_nm'].split()[0] : data[i]['tts'] for i in range(len(data))}
# ctr_cur = json.dumps(ctr_cur, indent = 4, ensure_ascii=False) 

# 전체 표 보여주기 -> 표 정리하고 싶음..ㅎ
summary = pd.DataFrame(data)
summary
```

```python
country_info = {data[i]['cur_nm'].split()[0]:data[i]['cur_nm'].split()[-1] for i in range(len(data))} # 나라이름, 단위
receive = { country_info[i][0] : data[i]['ttb'] for i in range(len(data))} # 나라이름 : ttb

send = { country_info[i][0] : data[i]['tts'] for i in range(len(data))}
send
```

```python
# User input
# 중국의 경우 '위안화' 로 검색, 유럽지역은 '유로'로 검색 , 덴마크는 덴마아크로 검색

mode = input('What do you want? (receive foreign money :1 / send foreign money :2)') #nlp 적용 가능할지도?
country = input('Which country or currency are you going to 거래? ')  # 국가/ 단위 둘다 접근 가능하게 처리 해야겠다.
    
if mode == 1: # receive : ttb
    money = input('How much money are you going to receive? ')
    print()
    r_money = receive[country]*money
    print(f'Receive : {r_money}{country_info[country][-1]}')
elif mode == 2:
    money = input('How much money are you going to send? ')
    print()
    r_money = send[country]*money
    print(f'Send : {r_money}won')


```
